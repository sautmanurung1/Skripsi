```python
import pandas as pd

data = pd.read_csv('Datas_Nilai.csv', sep=';')
data
```

```python
import math
import pandas as pd
import numpy as np
from collections import Counter

data_sem1_peng = data['Peng Sem 1'].astype(str)
data_sem1_ket = data['Ket Sem 1'].astype(str)
data_sem2_peng = data['Peng Sem 2'].astype(str)
data_sem2_ket = data['Ket Sem 2'].astype(str)

random_choices_sem1_peng = np.random.choice(data_sem1_peng, size=6)
random_choices_sem1_ket = np.random.choice(data_sem1_ket, size=6)
random_choices_sem2_peng = np.random.choice(data_sem2_peng, size=6)
random_choices_sem2_ket = np.random.choice(data_sem2_ket, size=6)

C1_1 = float(random_choices_sem1_peng[0].replace(',', '.'))
C2_1 = float(random_choices_sem1_peng[1].replace(',', '.'))
C3_1 = float(random_choices_sem1_peng[2].replace(',', '.'))
C1_1_2 = float(random_choices_sem1_peng[3].replace(',', '.'))
C2_1_2 = float(random_choices_sem1_peng[4].replace(',', '.'))
C3_1_2 = float(random_choices_sem1_peng[5].replace(',', '.'))
C1_2 = float(random_choices_sem1_ket[0].replace(',', '.'))
C2_2 = float(random_choices_sem1_ket[1].replace(',', '.'))
C3_2 = float(random_choices_sem1_ket[2].replace(',', '.'))
C1_2_2 = float(random_choices_sem1_ket[3].replace(',', '.'))
C2_2_2 = float(random_choices_sem1_ket[4].replace(',', '.'))
C3_2_2 = float(random_choices_sem1_ket[5].replace(',', '.'))
C1_3 = float(random_choices_sem2_peng[0].replace(',', '.'))
C2_3 = float(random_choices_sem2_peng[1].replace(',', '.'))
C3_3 = float(random_choices_sem2_peng[2].replace(',', '.'))
C1_3_2 = float(random_choices_sem2_peng[3].replace(',', '.'))
C2_3_2 = float(random_choices_sem2_peng[4].replace(',', '.'))
C3_3_2 = float(random_choices_sem2_peng[5].replace(',', '.'))
C1_4 = float(random_choices_sem2_ket[0].replace(',', '.'))
C2_4 = float(random_choices_sem2_ket[1].replace(',', '.'))
C3_4 = float(random_choices_sem2_ket[2].replace(',', '.'))
C1_4_2 = float(random_choices_sem2_ket[3].replace(',', '.'))
C2_4_2 = float(random_choices_sem2_ket[4].replace(',', '.'))
C3_4_2 = float(random_choices_sem2_ket[5].replace(',', '.'))

result_c1 = []
result_c2 = []
result_c3 = []

result_c1_2 = []
result_c2_2 = []
result_c3_2 = []

result_categories = []
result_categories_2 = []

attempt = 0  
while attempt < 100:
    for i, row in data.iterrows():
        Peng_Sem1 = float(row['Peng Sem 1'].replace(',', '.'))
        Peng_Sem2 = float(row['Peng Sem 2'].replace(',', '.'))
        Ket_Sem1 = float(row['Ket Sem 1'].replace(',', '.'))
        Ket_Sem2 = float(row['Ket Sem 2'].replace(',', '.'))

        distance_c1 = math.sqrt((C1_1 - Peng_Sem1) ** 2 + (C1_2 - Ket_Sem1) ** 2 + (C1_3 - Peng_Sem2) ** 2 + (C1_4 - Ket_Sem2) ** 2)
        result_c1.append(distance_c1)

        distance_c2 = math.sqrt((C2_1 - Peng_Sem1) ** 2 + (C2_2 - Ket_Sem1) ** 2 + (C2_3 - Peng_Sem2) ** 2 + (C2_4 - Ket_Sem2) ** 2)
        result_c2.append(distance_c2)

        distance_c3 = math.sqrt((C3_1 - Peng_Sem1) ** 2 + (C3_2 - Ket_Sem1) ** 2 + (C3_3 - Peng_Sem2) ** 2 + (C3_4 - Ket_Sem2) ** 2)
        result_c3.append(distance_c3)

        distance_c1_2 = math.sqrt((C1_1_2 - Peng_Sem1) ** 2 + (C1_2_2 - Ket_Sem1) ** 2 + (C1_3_2 - Peng_Sem2) ** 2 + (C1_4_2 - Ket_Sem2) ** 2)
        result_c1_2.append(distance_c1_2)

        distance_c2_2 = math.sqrt((C2_1_2 - Peng_Sem1) ** 2 + (C2_2_2 - Ket_Sem1) ** 2 + (C2_3_2 - Peng_Sem2) ** 2 + (C2_4_2 - Ket_Sem2) ** 2)
        result_c2_2.append(distance_c2_2)

        distance_c3_2 = math.sqrt((C3_1_2 - Peng_Sem1) ** 2 + (C3_2_2 - Ket_Sem1) ** 2 + (C3_3_2 - Peng_Sem2) ** 2 + (C3_4_2 - Ket_Sem2) ** 2)
        result_c3_2.append(distance_c3_2)

        if distance_c1 < distance_c2 and distance_c1 < distance_c3:
            result_categories.append(1)
        elif distance_c2 < distance_c1 and distance_c2 < distance_c3:
            result_categories.append(2)
        else:
            result_categories.append(3)

        if distance_c1_2 < distance_c2_2 and distance_c1_2 < distance_c3_2:
            result_categories_2.append(1)
        elif distance_c2_2 < distance_c1_2 and distance_c2_2 < distance_c3_2:
            result_categories_2.append(2)
        else:
            result_categories_2.append(3)

    min_distances = [min(distances) for distances in zip(result_c1, result_c2, result_c3)]
    sum_distances = sum(min_distances)
    print("Min Distances:", min_distances)
    print("Sum Distances:", sum_distances)

    min_distances_2 = [min(distances) for distances in zip(result_c1_2, result_c2_2, result_c3_2)]
    sum_distances_2 = sum(min_distances_2)
    print("Min Distances 2:", min_distances_2)
    print("Sum Distances 2:", sum_distances_2)

    difference = sum_distances_2 - sum_distances
    print("Difference:", difference)

    count_category_1 = result_categories.count(1)
    print("Count Category 1:", count_category_1)

    count_category_2 = result_categories.count(2)
    print("Count Category 2:", count_category_2)

    count_category_3 = result_categories.count(3)
    print("Count Category 3:", count_category_3)

    if difference > 0:
        break

    result_c1.clear()
    result_c2.clear()
    result_c3.clear()
    result_c1_2.clear()
    result_c2_2.clear()
    result_c3_2.clear()
    result_categories.clear()
    result_categories_2.clear()

    attempt += 1

if attempt == 100 and difference <= 0:
    print("Positive difference not found after 100 attempts.")
```
