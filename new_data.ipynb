```python
import csv

# Specify the input CSV file
input_file = 'Data Siswa.csv'

# Specify the output CSV file
output_file = 'Data_siswa.csv'

# Open the input and output files
with open(input_file, 'r') as file_in, open(output_file, 'w', newline='') as file_out:
    # Create a CSV reader and writer objects
    reader = csv.reader(file_in, delimiter=';')
    writer = csv.writer(file_out)

    # Iterate over each row in the input file
    for row in reader:
        # Remove semicolons from each field in the row
        cleaned_row = [field.replace(';', '') for field in row]
        
        # Write the cleaned row to the output file
        writer.writerow(cleaned_row)
```
